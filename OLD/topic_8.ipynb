```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import json
```

```python
res_e = pd.read_csv('results/res_e.csv')
res_e.head()
```

```python
def dart(date):
    if date.split('.')[-1] in ['1','2','3','4','5','01','02','03','04','05'] :
        year = 2022
    else:
        year = 2021
    return f'{date.split(".")[0]}/{date.split(".")[1]}/{year}'
```

```python
e_p = pd.read_csv('plans/8_plan.csv', dtype={'дата': str})
e_p.dropna(inplace=True)
e_p[:15]
```

```python


e_p['дата'] = [dart(x) for x in e_p['дата']]
e_p['дата']=pd.to_datetime(e_p['дата'],format='%d/%m/%Y')

e_p['Тема уроку']=[x+' ' for x in e_p['Тема уроку']]
e_p = e_p.groupby('дата').sum()
e_p.to_csv('plans/8_plan_1.csv')
e_p[:30]
```

```python
res_e['date']=[e_p.iloc[e_p.index.get_loc(x,method='nearest')][-((index+1)%2)] for index,x in enumerate(res_e['date_'])]
res_e.to_csv('results/res_e_1.csv', index=False)
res_e
```
