```python
import os
import csv
from notion_client import Client
import pandas as pd

# Initialize Notion API client
notion = Client(auth=os.environ['NOTION_TOKEN'])
```

```python
# Replace 'your_database_id' with the ID of your database
database_id = os.environ['NOTION_DATABASE_ID']

# Query the database
results = notion.databases.query(database_id=database_id)
```

```python
def extract_data_and_export_to_csv(results):
    # Extract the data
    data = []
    columns = results["results"][0]["properties"].keys()

    for page in results["results"]:
        row = {}
        row['id'] = page["id"]
        for column in columns:
            property_type = page["properties"][column]["type"]
            if property_type in page["properties"][column]:
                row[column] = page["properties"][column][property_type]
            else:
                row[column] = ""
        data.append(row)

    # Write the data to a CSV file
    df = pd.DataFrame(data)
    df.to_csv("notion_database.csv", index=False, quoting=csv.QUOTE_NONNUMERIC)
    return df
```

```python
df = extract_data_and_export_to_csv(results)
df.head()
```
